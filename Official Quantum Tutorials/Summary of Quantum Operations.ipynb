```python
#useful additional packages
import matplotlib
import numpy as py
from math import pi
from ipynb.fs.full.RUN import run_program
```

```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile, execute
from qiskit.tools.visualization import circuit_drawer, plot_bloch_multivector
from qiskit.quantum_info import state_fidelity
from qiskit import BasicAer, Aer

backend = BasicAer.get_backend('unitary_simulator')
```

```python
#SINGLE QUBIT GATES
q = QuantumRegister(1)
```

```python
#U gates
qc = QuantumCircuit(q)
qc.u3(pi/2, pi/2, pi/2, q)
qc.draw()
job = backend.run(transpile(qc, backend))
job.result().get_unitary(qc, decimals = 3)

statevec = execute(qc, backend = Aer.get_backend("statevector_simulator")).result().get_statevector()
plot_bloch_multivector(statevec)
```
